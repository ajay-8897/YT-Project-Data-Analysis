2. Country-Level Insights
- Country Popularity: Aggregate subscribers/views by country.
- Population vs Subscribers: Normalize subscribers by population to see penetration rates.
- Unemployment vs Engagement: Explore correlations between unemployment rate and video consumption.
- Urban Population Influence: Higher urbanization may correlate with higher YouTube adoption.


```python
#  Country Popularity

country_stats = df.groupby('Country').agg({
    'subscribers':'sum',
    'video views':'sum'
}).sort_values('subscribers', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=country_stats.index, y=country_stats['subscribers'], color='skyblue')
plt.title("Top 10 Countries by Total Subscribers")
plt.xticks(rotation=45)
plt.ylabel("Subscribers")
plt.show()

# Top 10 countries by total subscribers
```

This code produces a bar chart showing the top 10 countries ranked by total YouTube subscribers.
- The dataset is grouped by Country, summing up all subscribers from channels in each nation.
- The countries are then sorted in descending order, so the chart highlights the biggest subscriber bases globally.
- The x-axis lists the top 10 countries, while the y-axis shows their total subscriber counts.
- Each bar is colored sky blue, making the comparison clear and visually consistent.
- The chart title “Top 10 Countries by Total Subscribers” explains the focus of the visualization.

```python

#  Population vs Subscribers (Penetration Rate)

df['subs_per_capita'] = df['subscribers'] / df['Population']

country_penetration = df.groupby('Country')['subs_per_capita'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=country_penetration.index, y=country_penetration.values, color='green')
plt.title("Top 10 Countries by Subscriber Penetration (Subscribers per Capita)")
plt.xticks(rotation=45)
plt.ylabel("Subscribers per Person")
plt.show()


# top 10 countries by subscriber penetration rate (subscribers per capita)
```

This code produces a bar chart showing the top 10 countries ranked by subscriber penetration (subscribers per capita).
- First, it calculates subs_per_capita by dividing each channel’s subscribers by the country’s population.
- Then, it groups by Country and takes the average penetration rate, sorting to find the top 10.
- The x-axis lists those countries, while the y-axis shows their average subscribers per person.
- Each bar is colored green, making the comparison clear.
- The chart title “Top 10 Countries by Subscriber Penetration (Subscribers per Capita)” explains the focus.

```python
#  Regional Dominance in Categories

country_category = df.groupby(['Country','category'])['subscribers'].sum().reset_index()

g = sns.catplot(
    data=country_category,
    x='Country', y='subscribers',
    kind='bar',
    col='category',
    col_wrap=3,
    height=5, aspect=1.2
)

g.set_titles("Category: {col_name}")
g.set_axis_labels("", "Total Subscribers")   # remove x-axis label

# Hide country names on x-axis
for ax in g.axes.flatten():
    ax.set_xticklabels([])   # remove tick labels
    ax.tick_params(axis='x', rotation=0)  # no rotation needed now

plt.subplots_adjust(top=0.9)
g.fig.suptitle("Regional Dominance in Categories", fontsize=12)
plt.show()

# Regional Dominance in Categories(Subscribers by Country & Category)
```

This code generates a multi-panel bar chart that breaks down YouTube subscriber dominance by country and category:
- The dataset is grouped by Country and category, summing total subscribers for each combination.
- Using Seaborn’s catplot, each category gets its own subplot (faceted into columns).
- Within each subplot, bars show the total subscribers per country for that category.
- col_wrap=3 ensures the plots wrap neatly into rows, three per row.
- Axis labels and rotated ticks make the country names readable.
- A shared suptitle ties all the subplots together: “Regional Dominance in Categories.”


4. Temporal Analysis
- Launch Year Trends: Which years produced the most successful channels?
- Seasonality: Check if certain months (Month_Number) correlate with higher channel creation.
- Longevity: Compare older channels (2006–2010) vs newer ones (2015–2020).


```python
#Launch Year Trends

year_stats = df.groupby('Year').agg({
    'Youtuber':'count',
    'subscribers':'sum',
    'video views':'sum'
}).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=year_stats, x='Year', y='Youtuber', color='skyblue')
plt.title("Number of Channels Launched per Year")
plt.xticks(rotation=45)
plt.ylabel("Channel Count")
plt.show()

plt.figure(figsize=(10,6))
sns.lineplot(data=year_stats, x='Year', y='subscribers', marker='o', color='green')
plt.title("Total Subscribers by Launch Year")
plt.ylabel("Subscribers")
plt.show()

plt.figure(figsize=(10,6))
sns.lineplot(data=year_stats, x='Year', y='video views', marker='o', color='red')
plt.title("Total Video Views by Launch Year")
plt.ylabel("Video Views")
plt.show()

# Launch Year Trends(Channel Count, Subscribers, Views over Years)
```

This block of code performs a temporal analysis of YouTube channel launches and visualizes three different aspects:
- Bar Chart (Channels Launched per Year):
Shows how many channels were created each year. Taller bars indicate years with more new channels.



- Line Chart (Subscribers by Launch Year):
Plots the total subscribers gained by channels launched in each year. The green line with markers highlights growth trends and peak years.


- Line Chart (Video Views by Launch Year):
Displays the total video views accumulated by channels launched in each year. The red line shows how viewership varies depending on launch year.

```python
#  Longevity (Older vs Newer Channels)

older = df[(df['Year'] >= 2006) & (df['Year'] <= 2010)]
newer = df[(df['Year'] >= 2015) & (df['Year'] <= 2020)]

longevity_stats = pd.DataFrame({
    'Period': ['2006-2010','2015-2020'],
    'Subscribers': [older['subscribers'].sum(), newer['subscribers'].sum()],
    'Views': [older['video views'].sum(), newer['video views'].sum()],
    'Channels': [older['Youtuber'].count(), newer['Youtuber'].count()]
})

longevity_stats.set_index('Period')[['Subscribers','Views','Channels']].plot(kind='bar', figsize=(10,6))
plt.title("Longevity Comparison: Older vs Newer Channels")
plt.ylabel("Totals")
plt.show()
```

This code creates a bar chart comparing older vs newer YouTube channels in terms of subscribers, views, and channel counts:
- The dataset is split into two periods: 2006–2010 (older channels) and 2015–2020 (newer channels).
- For each period, totals are calculated:
- Subscribers → sum of all subscribers.
- Views → sum of all video views.
- Channels → number of channels created.
- These values are stored in a DataFrame (longevity_stats) and plotted as grouped bars.
- The chart title “Longevity Comparison: Older vs Newer Channels” clarifies the focus.
